```python
data_df
```

```python
def clean_location(data_df):
    # "°"와 "N"을 빈 문자열로 대체하여 제거
    cleaned_location = data_df.replace(" N", "")
    return cleaned_location
# 'Location' 컬럼에 apply() 함수 적용
data_df['LAT'] = data_df['LAT'].apply(clean_location)
```
