```python
import pandas as pd

data = pd.read_csv('datasets/국내지진목록 5년차.csv')
data
```

```python
data['OT'] = pd.to_datetime(data['OT']).dt.strftime('%Y-%m-%d')
data
```

```python
def clean_location(data):
    # "°"와 "N"을 빈 문자열로 대체하여 제거
    cleaned_location = data.replace(" N", "")
    return cleaned_location
# 'Location' 컬럼에 apply() 함수 적용
data['LAT'] = data['LAT'].apply(clean_location)

def clean_location(data):
    # "°"와 "N"을 빈 문자열로 대체하여 제거
    cleaned_location = data.replace(" E", "")
    return cleaned_location
# 'Location' 컬럼에 apply() 함수 적용
data['LON'] = data['LON'].apply(clean_location)
```

```python
data
```

```python
data['SC'] = data['SC'].astype(str)
```

```python
# 로마숫자를 아라비아 숫자로
def hihi(a):
    if a == "Ⅰ":
        return 1
    elif a == "Ⅱ":
        return 2
    elif a == "Ⅲ":
        return 3
    elif a == "Ⅳ":
        return 4
    else:
        return 5
data['SC'] = data['SC'].apply(hihi)
```
